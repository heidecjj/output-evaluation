# This document contains examples of creating plots from data

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
dfConst = pd.read_csv('../datasets/constantOutput.csv',
                 names=['time', 'v1', 'v2'])
dfAdapt = pd.read_csv('../datasets/adaptiveOutput.csv',
                 names=['time', 'v1', 'v2'])

dfConst.head()
```

## Single plot

```python
dfConst['v1'].plot(alpha=0.6)
plt.xlabel('time (us)')
plt.ylabel('membrane potential (V)')
plt.title('simple-synapse configuration')
```

## Multiple plots

```python
dfConst[['v1', 'v2']].plot(alpha=0.6)
```

```python
dfConst['v1'].plot(alpha=0.6)
dfAdapt['v1'].plot(alpha=0.6)
```

## High Quality rendering for presentations

```python
# Select desired indices
timeIndices = dfAdapt['time'] < 100

# Select color, figure size (in inches), alpha, and legend label
# Also 
dfAdapt[['time','v1']][timeIndices].plot(x='time', y='v1', alpha=0.6, 
                                    color='red',
                                    figsize=(8, 6),
                                    label='v1 (adaptive time stepping)')

# Axis labelling
plt.xlabel('time (ms)')
plt.ylabel('membrane potential (mV)')

# Give path for save, set dpi (google appropriate DPI for 
# powerpoint size / poster size)
plt.savefig('temp.png', dpi=300)
```
