```python
import pandas as pd
import seaborn as sb
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('../datasets/adaptiveOutput.csv',
                 names=['time', 'v1', 'v2'])
df.head()
```

```python
df['dt'] = df['time'].diff()
df['dv1'] = df['v1'].diff() / df['dt']
df['dv2'] = df['v2'].diff() / df['dt']

df['ddv1'] = df['dv1'].diff() / df['dt']
df['ddv2'] = df['dv2'].diff() / df['dt']

# Throw away the first two rows bc it doesnt have a derivatives
df = df.drop(df.index[0:2])
```

```python
# Standardize
# df = (df - df.mean()) / df.std()
```

```python
vsPlots = sb.pairplot(df)
```
